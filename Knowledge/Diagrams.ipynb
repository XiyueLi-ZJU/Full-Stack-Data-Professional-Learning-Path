# Diagrams

A picture is worth a thousand words. Diagrams break down complex systems into smaller parts, visualize how various parts of a system interact with each other. Diagrams make it easier to communicate to different stakeholders, both technical and nontechnical. Diagrams also serve as an important documentation which can be crucial for troubleshooting, maintenance, improvement and knowledge sharing. A Data Professional needs to be familiar with various of kinds of diagrams. Below are some diagrams that a data professional needs to know.

# (1) Entity Relationship Diagram (ERD)

ERD visualizes how entities relate to each other within a system.There are three major components in an ERD: **Entity**, **Attribute**, and **Relationship**. Entities are real-world objects or concepts, attributes are properties or characteristics of entities, and relationships define how entities interact with each other. Based on the level of details, ERD can be further categorized into three types: Conceptual ERD, Logical ERD, and Physical ERD. Conceptual ERD focuses on high-level design, Logical ERD adds more details such as primary keys and foreign keys, Physical ERD depicts all details, inlucing table names, column names, column data types, constraints, etc.  ERD is commonly used in relational database design and management. Typically it is not used for unstructured data.

# (2) Data Flow Diagram (DFD)

DFD provides a graphical representation of the flow of data through a system. It depicts how data is inputted, processed, distributed, and stored in a system. It is commonly used for system design. A DFD has a set of predefined symbols and notations: **External Entity**, **Data Store**, **Process**, and **Data Flow**. External Entity is the source and destination of data, Data Store represents where data is held for future use, Process is the procedure that changes the data, Data Flow shows the movement of data. There are two types of DFD: [Logical DFD and Physical DFD](https://www.lucidchart.com/pages/data-flow-diagram/logical-vs-physical-data-flow-diagram?gclid=Cj0KCQjw8NilBhDOARIsAHzpbLAT2TrqfXt83Pj8AhIw7KeYOdRiZbFwqNyuZEblJsMqqpdA-ygDXjsaApLCEALw_wcB&km_CPC_AdGroupID=147236965677&km_CPC_AdPosition=&km_CPC_CampaignId=19866748684&km_CPC_Country=1006912&km_CPC_Creative=651989681439&km_CPC_Device=c&km_CPC_ExtensionID=&km_CPC_Keyword=workflow%20diagram&km_CPC_MatchType=e&km_CPC_Network=g&km_CPC_TargetID=kwd-26222580&km_CPC_placement=&km_CPC_target=&utm_campaign=_chart_en_tier1_desktop_search_strategic_exact_&utm_medium=cpc&utm_source=google). Logical DFD describes what happens in a system while Physical DFD describes how it happens.

# (3) Data Lineage Diagram

Data Lineage Diagram visualizes the life-cycle of data, from its origin to destination, and describes how the data is transformed and used. Therefore it is useful for troubleshooting and data governance. Data Lineage Diagrams can be presented at different granularity levels, from system level to column level. If the data platform you are using has a Data Lineage module, you may get a Data Lineage Diagram generated by the data platform. You can also create the data lineage diagram yourself using some external visualization tools such as miro, draw.io, and Lucidchart etc. 

For more detailed information and examples 
- [Data lineage in Microsoft Purview](https://learn.microsoft.com/en-us/purview/concept-data-lineage)
- [Google Data Catalog - Data lineage](https://cloud.google.com/data-catalog/docs/concepts/about-data-lineage)

# (4) Flow Chart

Flow charts visualize workflow and processes. It can be used for various purposes. The basic flow chart, Workflow Diagram, focuses on process mapping. Swimlane Diagram, a type of flow chart, focuses more on who does what and is commonly used for task assignment and job responsibilities clarification.
